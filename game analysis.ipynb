```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:/Users/Aluno/VendaDeJogos.csv', engine='python') #Comando usado para ler o arquivo
print (df)
```

```python
df = df.rename(columns= {'Rank': 'Mais vendidos',
                    'Name':'Nome do jogo',
                    'Year':'Ano',
                    'Platform':'Plataforma',
                    'Genre':'Gênero',
                    'Publisher':'Desenvolvedora',
                    'NA_Sales':'Vendas na América do norte',
                    'EU_Sales':'Vendas na Europa',
                    'JP_Sales':'Vendas no Japão',
                    'Other_Sales':'Outros locais','Global_Sales':'Venda Global'}) 
#Renomeação do nome das colunas
print (df)
```

```python
df.drop_duplicates() #Remoção das informações duplas
```

```python
df.dropna() #Remoção das informações NA
```

```python
df.['Gênero'].value_counts('Sports')*100
```

```python
print('Gráfico das desenvolvedoras com mais jogos na lista de TOP 100')
df["Gênero"].value_counts().plot.pie(figsize=(8,8)) # Este gráfico representa as informações acima
```

```python
df.head(n=100)['Desenvolvedora'].value_counts() #Porcentagem das desenvolvedoras com mais jogos na lista
#Analisando os 100 primeiros jogos, podemos observar que a Nintendo está em mais da metade deles
```

```python
print ('Porcentagem das desenvolvedoras com mais jogos na lista')
df.head(n=100)['Desenvolvedora'].value_counts().plot.barh(figsize=(12,8)) #Gráfico representando as informações acima
```
